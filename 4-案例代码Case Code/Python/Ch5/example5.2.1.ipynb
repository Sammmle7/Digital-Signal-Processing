采用直接I型实现系统函数如下的IIR数字滤波器，并求单位脉冲响应和单位阶跃响应：

$H(z)=\frac{\sum_{r=0}^{M}b_{r}z^{-r}}{\sum_{k=0}^{N}a_{k}z^{-k}}=\frac{1-3z^{-1}+11z^{-2}+27z^{-3}+18z^{-4}}{16+12z^{-1}+2z^{-2}-4z^{-3}-2z^{-4}}$

```python
#导入使用的库
import numpy as np
import matplotlib.pyplot as plt
from scipy import signal
```

```python
#差分方程参数设置
b = np.array([1,-3,11,27,18]) #分子参数
a = np.array([16,12,2,-4,-2]) #分母参数
```

```python
#输入信号
N = 30
delta = signal.unit_impulse(N) #单位采样信号
y = np.ones(N) #单位阶跃信号
```

```python
#IIR 滤波器
zi = signal.lfilter_zi(b,a)*0 #零状态条件
z1,_ = signal.lfilter(b,a,delta,zi=zi);z2,_ =signal.lfilter(b,a,y,zi=zi)
```

单位脉冲响应和单位阶跃响应如下：

```python
#画图
fig,axs = plt.subplots(2,1,constrained_layout=True)
axs[0].stem(z1,basefmt="");axs[1].stem(z2,basefmt="")
plt.rcParams['font.sans-serif']=['SimHei'] #用来正常显示中文标签
axs[0].set_title('IIR直接型h(n)');axs[1].set_title('IIR直接型y(n)')
plt.show();fig.savefig('./iir_dir_I_sequence.png',dpi=500)
```
