1.混叠现象

```python
#导入使用的库
import numpy as np
import pylab as pl
```

```python
fs = 900 #自己改频率
N = 512
t = np.arange(0, 1, 1/fs)
x = np.sin(2*np.pi*100*t)+2*np.sin(2*np.pi*400*t)
```

```python
xs = x[:N]
xf = np.fft.rfft(xs)/N
freqs = np.linspace(0, int(fs/2), int(N/2+1))
xfp = np.abs(xf)
pl.figure()
pl.subplot(211)
pl.plot(t[:N], xs)
pl.xlabel(u"Time(S)")
pl.subplot(212)
pl.plot(freqs, xfp)
pl.xlabel(u"Freq(Hz)")
pl.show()
```

2.栅栏效应

```python
#导入使用的库
import numpy as np
import pylab as pl
import matplotlib.pyplot as plt
```

```python
f = 8000
N = 512 #自己改N，相应长度也要改
t = np.arange(0, 1, 1/f)
x = np.sin(2*np.pi*100*t)  + 2*np.sin(2*np.pi*300*t)
```

```python
A = x[:N]
y = np.fft.rfft(A,N)/(N)
freqs = np.linspace(0, int(f/2), int(N/2+1))
B = np.abs(y)
plt.figure()
plt.subplot(211)
plt.plot(t[:N], A)
plt.xlabel("t/s")
plt.subplot(212)
plt.plot(freqs, B)
plt.xlabel("f(Hz)")
plt.show()
```
