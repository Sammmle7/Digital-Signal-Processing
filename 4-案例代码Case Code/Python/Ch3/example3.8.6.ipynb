最优化窗是指在在某种准则下得到的窗函数，下面展示的是高斯窗最优化窗的程序实现：

```python
#导入使用的库
import numpy as np
from scipy import signal,fft
import matplotlib.pyplot as plt
from matplotlib.ticker import MaxNLocator
```

```python
#计算高斯窗的wn值和幅度响应
N = 51 #高斯窗长度
wn = signal.windows.gaussian(N,7) #高斯窗的wn值
N0 = 2048;N1 = int(N0/2)
Ha = np.abs(fft.fft(wn,N0))+1e-10;Ha = Ha/np.max(Ha)
Ar = 20*np.log10(Ha);freq = np.linspace(0,1,N1)
```

当$1/a=7$时，高斯窗的$\omega(n)$和幅度响应绘制如下：

```python
#绘制高斯窗
fig,ax = plt.subplots();ax.stem(wn,basefmt="")
ax.set_title('高斯窗(a = 1/7)')
ax.set_xlabel('n')
ax.set_ylim([0,1.5])
plt.rcParams['font.sans-serif']=['SimHei'] #用来正常显示中文标签
fig.savefig('./win_gauss1.png',dpi=500)
```

```python
#绘制高斯窗的幅度响应
fig,ax = plt.subplots();ax.plot(freq,Ar[:N1]);ax.grid()
ax.set_title('高斯窗的幅度响应(a= 1/7)');ax.set_xlabel('k')
ax.set_xlabel(r'$ \omega / \pi $')
ax.set_ylabel(r'$ 20log_{10}| H ( \omega ) | $')
ax.set_xlim([0,1]);ax.set_ylim([-100,1])
ax.xaxis.set_major_locator(MaxNLocator(11))
ax.yaxis.set_major_locator(MaxNLocator(11))
plt.rcParams['axes.unicode_minus'] = False #用来显示负号
fig.savefig('./win_gauss2.png',dpi=500)
```
